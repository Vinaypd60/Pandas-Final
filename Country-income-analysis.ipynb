# Visualization - seaborn & matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = 6,2

import warnings
warnings.filterwarnings('ignore')
```

```python
df['InternetUsers']
```

```python
# Univariate analysis = plot the graph using one variable is called univariate analysis
vis1 = sns.distplot(df['InternetUsers'])
plt.show()
```

```python
vis1 = sns.distplot(df['InternetUsers'], bins=10)
plt.show()
```

```python
# Bivariant
vis2 = sns.boxplot(df,x='IncomeGroup',y='BirthRate')
plt.show()
```

```python
vis3 = sns.lmplot(df,x='InternetUsers',y='BirthRate',fit_reg=False)
plt.show()
```

```python
vis3 = sns.lmplot(df,x='InternetUsers',y='BirthRate',fit_reg=False)
plt.show()
```

```python
vis4 = sns.lmplot(df,x='InternetUsers',y='BirthRate', fit_reg=True, hue='IncomeGroup')
plt.show()
```
