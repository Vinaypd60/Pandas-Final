# What os Pandas?
## Pandas is the open-source library that is built on top of NumPy library.
## It is python package that offers various data structure.
## Python Provides the pandas module for high-performance data analysis.

# Why Do we need Pandas?
## It is a fast, flexible and powerful data manipulation library. Pandas is the most crucial module for applying DS using python.
## It provides several data structures like series, Index and DF for DA.

# Advantages of Pandas
## Easy import and export of data into a tabular format data structure like DF.
## Routined for manipulation and complex analysis of data.
## Easy handling of missing data(presented as NaN) in floating point as well as non-floating point data.
## DataSet cleaning, merging, and joining.
## Reshaping of datasets
## Columns can be inserted, deleted and updated from DF and higher dimensional objects.
## TimeSeries based data manipulation and analysis API.
## GroupBy functionality to perform split-apply combine operation.
## Well integrated with other libraries like NumPy and Matplotlib.
