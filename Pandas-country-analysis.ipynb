# Visualization: Matplotlib & Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = 6,2

import warnings
warnings.filterwarnings('ignore')
```

```python
data['InternetUsers']
```

```python
vis1 = sns.distplot(data['InternetUsers'])
plt.show()
```

```python
vis1 = sns.distplot(data['InternetUsers'],bins=10)
plt.show()
```

```python
vis2 = sns.distplot(data['BirthRate'])
plt.show()
```

```python
vis2 = sns.distplot(data['BirthRate'],bins=5)
plt.show()
```

```python
vis3 = sns.boxplot(data, x = 'IncomeGroup', y = 'BirthRate')
plt.show()
```

```python
vis4 = sns.lmplot(data, x = 'InternetUsers', y = 'BirthRate' )
plt.show()
```

```python
vis4 = sns.lmplot(data, x = 'InternetUsers', y = 'BirthRate', fit_reg=False)
plt.show()
```

```python
vis4 = sns.lmplot(data, x = 'InternetUsers', y = 'BirthRate', fit_reg=True, hue='IncomeGroup')
plt.show()
```
