```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Months':['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales':[10000,12000,15000,13000,17000,16000],
    'Profit':[2000,2500,3000,2800,3500,3200]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5)) # here 8 is width and 5 is height
plt.plot(df['Months'],df['Sales'],color = 'Blue', marker='o',linestyle='--', label='Sales')
plt.title('Sales Trend Over Months')
plt.xlabel('Months')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6)) # here 8 is width and 5 is height
width = 0.3 
plt.bar(df['Months'],df['Sales'], width=width,color = 'skyblue', label='Sales')
plt.bar(df['Months'],df['Profit'], width=width,color = 'green', label='Profit',bottom=df['Sales'])
plt.title('Sales & Profit comparison by Month')
plt.xlabel('Months')
plt.ylabel('Amount ($)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.pie(df['Profit'], labels=df['Months'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Profit Distribution by Month')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df['Sales'], df['Profit'], color='green', s=100, edgecolors='black')
plt.title('Sales vs Profit')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt

data = {
      'Months': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
      'Sales': [10000, 12000, 15000, 13000, 17000, 16000],
      'Profit': [2000, 2500, 3000, 2800, 3500, 3200]
}
df = pd.DataFrame(data)

def generate_plot(plot_type):
    plt.figure(figsize=(8, 5))
   
    if plot_type == 'Line Plot':
        plt.plot(df['Months'], df['Sales'], color='blue', marker='o', linestyle='--', label='Sales')
        plt.title('Sales Trend Over Months')
        plt.xlabel('Months')
        plt.ylabel('Sales ($)')
        plt.grid(True)
        plt.legend()
       
    elif plot_type == 'Bar Chart':
        width = 0.3
        plt.bar(df['Months'], df['Sales'], width=width, label='Sales', color='skyblue')
        plt.bar(df['Months'], df['Profit'], width=width, label='Profit', color='orange', bottom=df['Sales'])
        plt.title('Sales and Profit comparison by Month')
        plt.xlabel('Months')
        plt.ylabel('Amount ($)')
        plt.legend()
       
    elif plot_type == 'Pie Chart':
        plt.pie(df['Profit'], labels=df['Months'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
        plt.title('Profit Distribution by Month')
       
    elif plot_type == 'Scatter Plot':
        plt.scatter(df['Sales'], df['Profit'], color='green', s=100, edgecolors='black')
        plt.title('Sales vs Profit')
        plt.xlabel('Sales ($)')
        plt.ylabel('Profit ($)')
        plt.grid(True)
   
    plt.tight_layout()
    return plt


demo = gr.Interface(
    fn=generate_plot,
    inputs=gr.Dropdown(choices=['Line Plot', 'Bar Chart', 'Pie Chart', 'Scatter Plot', 'Box Plot'], label='Select Plot Type'),
    outputs=gr.Plot(label='Generated Plot'),
    title='Dynamic Data Visualization with Gradio',
    description='Select a plot type to visualize the sales and profit data.'
)

demo.launch()
```
